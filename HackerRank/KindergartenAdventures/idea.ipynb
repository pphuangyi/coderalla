```python
import os
import sys


def solve(times):
    T = len(times)
    
    # Preprocessing
    # s for start, e for end, both inclusive
    sCounts, eCounts = [0] * T, [0] * T
    for i, t in enumerate(times):
        # If a student don't need extra time or takes too long to finish
        # We don't have to consider the student. 
        if t >= T or t <= 0:
            continue
        
        s, e = (T - i) % T, (T - i + (t - 1)) % T
        sCounts[s] += 1
        eCounts[e] += 1
        if e < s:
            sCounts[0] += 1 
            eCounts[T - 1] += 1

    sCumsums, eCumsums = [sCounts[0]], [eCounts[0]]
    for i in range(1, T):
        sc, ec = sCumsums[-1] + sCounts[i], eCumsums[-1] + eCounts[i]
        sCumsums.append(sc)
        eCumsums.append(ec)
    
    # print(sCumsums)
    # print(eCumsums)
        
    # Get the best position to start collection
    min_missing, min_pos = float('inf'), 0
    for i in range(T):
        sc, ec = sCumsums[i], 0
        if i > 0:
            ec = eCumsums[i - 1]
        missing = sc - ec
        if min_missing > missing:
            min_missing = missing
            min_pos = i
    
    return min_pos + 1
        
        

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t_count = int(input())
    t = list(map(int, input().rstrip().split()))
    result = solve(t)
    
    fptr.write(f'{result}\n')
    fptr.close()
```
