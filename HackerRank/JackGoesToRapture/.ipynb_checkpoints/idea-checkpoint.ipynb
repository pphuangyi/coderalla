```python
def getAdj(edges):
    adj = {}
    for h, t, fare in edges:
        if h in adj:
            adj[h][t] = fare
        else:
            adj[h] = {t: fare}
        if t in adj:
            adj[t][h] = fare
        else:
            adj[t] = {h: fare}
            
    return adj


def getCost(num_nodes, edges):

    adj = getAdj(edges)
    
    dist = {i + 1: float('inf') for i in range(num_nodes)}
    dist[1] = 0
    
    while len(dist) > 0:
        min_distance, min_node = float('inf'), None
        for n, d in dist.items():
            if min_distance > d:
                min_node = n
                min_distance = d
        
        if min_node == num_nodes or min_distance == float('inf'):
            break
        
        del dist[min_node]
        print(f'{min_node} {min_distance}')
        
        for neighbor in adj[min_node]:
            if neighbor in dist:
                distance = max(min_distance, adj[min_node][neighbor])
                if distance < dist[neighbor]:
                    dist[neighbor] = distance
        
    
    if min_distance < float('inf'):
        return min_distance
    else:
        return 'NO PATH EXISTS'


getCost(num_nodes, edges)
```

```python
input_index = '15'
input_fname = f'data/input{input_index}.txt'
with open(input_fname, 'r') as handle:
    num_nodes, num_edges = list(map(int, handle.readline().strip().split()))
    print(num_edges)
    edges = [list(map(int, handle.readline().strip().split())) for i in range(num_edges)]

getCost(num_nodes, edges)
```
