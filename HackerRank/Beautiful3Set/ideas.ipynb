```python
def find_k(n):
    if n == 1:
        return [[0, 0, 1]], 1
    # case 1: 2k
    if (n + 2) % 3 == 0:
        k1 = (n + 2) // 3 - 1
    else:
        k1 = (n + 2) // 3

    if (n - k1 - 1) % 2 == 0:
        k1 -= 1

    M1 = (n + k1) // 2

    # case 2: 2k + 1
    if (n - 1) % 3 == 0:
        k2 = (n - 1) // 3 - 1
    else:
        k2 = (n - 1) // 3

    if (n - k2) % 2 == 0:
        k2 -= 1

    M2 = (n + k2 + 1) // 2
    
    if M1 > M2:
        A = [[i, 2 * (k1 - i), n - 2 * k1 + i] for i in range(k1 + 1)]
        B = [[k1 + 1 + i, n - k1 - 1 - 2 * i, i] for i in range((n - k1) // 2)]
#         print(f'{n}: {k1}, {M1 + 1}')
#         print(f'\t|A| = {len(A)}')
#         print(f'\t|B| = {len(B)}')

        return A + B, M1 + 1
    else:
        A = [[i, 2 * (k2 - i) + 1, n - 1 - 2 * k2 + i] for i in range(k2 + 1)]
        B = [[k2 + 1 + i, n - k2 - 1 - 2 * i, i] for i in range((n - k2 + 1) // 2)]
#         print(f'{n}: {k2}, {M2 + 1}')
#         print(f'\t|A| = {len(A)}')
#         print(f'\t|B| = {len(B)}')
        return A + B, M2 + 1
```

```python
for n in range(300):
    S, m = find_k(n + 1)
    print(f'\t{n + 1}: {m}')#, {S}')
```
