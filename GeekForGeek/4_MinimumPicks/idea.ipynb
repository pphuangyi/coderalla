## accurate (but inefficient) solution

```python
def rec(arr, total):
    if sum(arr) < total:
        return None
    
    if len(arr) == 1:
        return [[total]]

    results = []
    for i in range(min(total, arr[-1]) + 1):
        tmp = rec(arr[:-1], total - i)
        if tmp is not None:
            results += [t + [i] for t in tmp]
            
    return results


def get_worst(picks):
    
    min_pairs = None
    worse_picks = None
    for r in picks:
        pairs = sum([a // 2 for a in r])
        if min_pairs is None or min_pairs >= pairs:
            min_pairs = pairs
            worst_pick = r

    return min_pairs, worst_pick
```

```python
n = random.randint(5, 20)
arr = [random.randint(5, 15) for _ in range(n)]
K = sum([a // 2 for a in arr])
k = random.randint((K - 10), K)

print(f'number of colors = {n}')
print(f'socks = {arr}')
print(f'pairs needed = {k}')

for total in range(2 * k, sum(arr) + 1):
    picks = rec(arr, total)
    min_pairs, _ = get_worst(picks)
    
    if min_pairs == k:
        break

print(f'total={total}, minimum pairs = {min_pairs}')
```

## More efficient solution

```python
def func(arr, k):
    brr = [a - 1 for a in arr]
    pairs = sum([b // 2 for b in brr])
    if pairs >= k:
        print('\tnormal case')
        solution = 2 * (k - 1) + len(brr) + 1
    else:
        print('\tnot normal case')
        diff = k - pairs
        solution = 2 * k + len(brr) - diff
    
    return solution
```

```python
num_tries = 100

for t in range(num_tries):
    n = random.randint(5, 8)
    arr = [random.randint(5, 8) for _ in range(n)]
    K = sum([a // 2 for a in arr])
    k = random.randint((K - 5), K)
    
    print(f'try {t}:')
    print(f'\tsocks = {arr}')
    print(f'\tpairs needed = {k}')
    
    for total in range(2 * k, sum(arr) + 1):
        picks = rec(arr, total)
        min_pairs, worst_pick = get_worst(picks)

        if min_pairs == k:
            break
            
    sol = func(arr, k)
    if sol != total:
        print(f'\tworst pick = {worst_pick}')
        print(f'\ttotal={total}, solution = {sol}')
```
