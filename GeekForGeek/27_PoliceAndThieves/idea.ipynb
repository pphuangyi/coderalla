# Police and Thieves
Given an array of size $n$ such that each element contains either a 'P' for policeman or a 'T' for thief. 
Find the maximum number of thieves that can be caught by the police. 

Keep in mind the following conditions:

- Each policeman can catch only one thief.
- A policeman cannot catch a thief who is more than $k$ units away from him.

```python
from bisect import bisect_left
```

```python
class Solution:
    def catchThieves(self, arr, n, k): 
        num_thieves_caught = 0
        idle_polices, uncaught_thieves = [], []
        for idx, figure in enumerate(arr):
            if figure == 'P':
                # Try to catch a thief
                j = bisect_left(uncaught_thieves, idx - k)
                if j < len(uncaught_thieves):
                    uncaught_thieves = uncaught_thieves[j + 1:]
                    num_thieves_caught += 1
                else:
                    # if there is no thief to be caught
                    idle_polices.append(idx)            
            else:
                # See whether there is an idle police
                j = bisect_left(idle_polices, idx - k)
                if j < len(idle_polices):
                    idle_polices = idle_polices[j + 1:]
                    num_thieves_caught += 1
                else:
                    # if there isn't an idle police to do the arrest
                    uncaught_thieves.append(idx)
                    
            print(f'{idx}, {figure}')
            print(f'\tidle polices = {idle_polices}, uncaught thieves = {uncaught_thieves}')
            
        return num_thieves_caught
```

```python
input_index = '01'
input_fname = f'data/input{input_index}.txt'
with open(input_fname, 'r') as handle:
    n, k = list(map(int, handle.readline().strip().split()))
    arr = list(handle.readline().strip().split())

print(f'n = {n}, k = {k}')  
end = len(arr)
Solution().catchThieves(arr[:end], end, k)
```
