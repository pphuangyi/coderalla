```python
from collections import deque
import matplotlib.pyplot as plt
def findTheWinner(n: int, k: int) -> int:
    memo = deque([i + 1 for i in range(n)])
    while len(memo) > 1:
        memo.rotate(-k)
        memo.pop()
    return memo.pop()
```

```python
n = 500
result = []
for k in range(1, n + 1):
    result.append(findTheWinner(n, k))

fig, ax = plt.subplots(1, 1, figsize=(10, 3))
ax.plot(range(1, n + 1), result)
```
