```python
def maxNumber(nums1, nums2, k):
    def best_k(nums, k):
        last, need, ans = -1, k, []
        while need > 0:
            val, idx = -1, last
            for i in range(last + 1, len(nums) - need + 1):
                if nums[i] > val:
                    val = nums[i]
                    idx = i
            ans.append(val)
            need -= 1
            last = idx
        return ans
    
    def merge(nums1, nums2):
        nums1 = ''.join(map(str, nums1))
        nums2 = ''.join(map(str, nums2))
        i, j, ans = 0, 0, ''
        while i < len(nums1) and j < len(nums2):
            if nums1[i] > nums2[j]:
                ans += nums1[i]
                i += 1
            elif nums1[i] < nums2[j]:
                ans += nums2[j]
                j += 1
            else:
                if nums1[i:] > nums2[j:]:
                    ans += nums1[i]
                    i += 1
                else:
                    ans += nums2[j]
                    j += 1
        return list(map(int, ans + nums1[i:] + nums2[j:]))

    biggest = [0] * k
    for k1 in range(max(0, k - len(nums2)), min(k, len(nums1)) + 1):
        k2 = k - k1
        number = merge(best_k(nums1, k1), best_k(nums2, k2))
        biggest = max(biggest, number)
    
    return biggest
```

```python
def merge(nums1, nums2):
    str1 = ''.join(map(str, nums1))
    str2 = ''.join(map(str, nums2))
    memo = [str2[i:] for i in range(len(nums2) + 1)]

    for i in range(len(nums1) - 1, -1, -1):
        memo_new = [''] * len(nums2) + [str1[i:]]
        for j in range(len(nums2) - 1, -1, -1):
            if nums1[i] > nums2[j]:
                memo_new[j] = str(nums1[i]) + memo[j]
            elif nums1[i] < nums2[j]:
                memo_new[j] = str(nums2[j]) + memo_new[j + 1]
            else:
                memo_new[j] = str(nums1[i]) + max(memo_new[j + 1], memo[j])
        memo = memo_new
    return list(map(int, memo[0]))

nums1 = [2,5,6,4,4,0]
nums2 = [7,3,8,0,6,5,7,6,2]

merge(nums1, nums2)
```
