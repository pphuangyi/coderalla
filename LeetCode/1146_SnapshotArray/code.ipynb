```python
def get_cmds_args(fname):
    with open(fname, 'r') as fh:
        cmds = fh.readline().strip()
        args = fh.readline().strip()

    cmds = cmds[2: -2].split('","')

    def get_args(args_str):
        if ',' in args_str:
            args = list(map(int, args_str.split(',')))
        else:
            try:
                args = [int(args_str)]
            except:
                args = []
        return args

    args = [get_args(args) for args in args[2: -2].split('],[')]

    return cmds, args
```

```python
from bisect import bisect_right
```

```python
class SnapshotArray:
    def __init__(self, length: int):
        self.length = length
        self.num_snaps = 0
        self.arr = [0] * self.length
        self.changed = []
        self.memo = [[[],[]] for i in range(self.length)]

    def set(self, index: int, val: int) -> None:
        self.arr[index] = val
        self.changed.append(index)
        

    def snap(self) -> int:
        for i in self.changed:
            self.memo[i][0].append(self.num_snaps)
            self.memo[i][1].append(self.arr[i])
        self.changed = []
            
        self.num_snaps += 1
        return self.num_snaps - 1


    def get(self, index: int, snap_id: int) -> int:
        idx = bisect_right(self.memo[index][0], snap_id)
        if idx == 0:
            return 0
        else:
            return self.memo[index][1][idx - 1]
```

```python
cmds, args = get_cmds_args('data/input_1')
```

```python
ssa = SnapshotArray(*args[0])
for cmd, arg in zip(cmds[1:], args[1:]):
    func = getattr(ssa, cmd)
    result = func(*arg)
    print(f'{cmd}({arg}) = {result}')
```
